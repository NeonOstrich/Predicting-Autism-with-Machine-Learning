```python
import pandas as pd
import os
import numpy as numpy

# File to Load (Remember to Change These)
train_data = "Data/train.csv"
```

```python
df_train_data = pd.read_csv(train_data)
```

```python
df_train_data[0:5]
```

```python
column_mapping = {'jaundice': 'jaundice_at_birth',
                  'autism': 'family_history_of_autism',
                   'relation': 'test_taker',
                  'result': 'ASQ_Total_score',
                 'contry_of_res': 'country_of_residence',
                 'Class/ASD': 'Autism_Diagnosis'}
```

```python
df_train_data = df_train_data.rename(columns=column_mapping)
```

```python
df_train_data = df_train_data.drop(['ID', 'age_desc'], axis=1)
```

```python
df_train_data[0:5]
```

```python
count = df_train_data.nunique()
print(count)
```

```python
count_type = df_train_data.country_of_residence.value_counts()
print(count_type)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
countries_to_replace = list(count_type[count_type<5].index)

# Replace in dataframe
for countries in countries_to_replace:
    df_train_data['country_of_residence'] = df_train_data['country_of_residence'].replace(countries,"Other")

# Check to make sure binning was successful
df_train_data['country_of_residence'].value_counts()
```

```python
train_dummies = pd.get_dummies(df_train_data)
train_dummies.head(5)
```

```python
train_dummies.dtypes
```

```python
train_dummies = train_dummies.astype('float')
```

```python
string_columns = train_dummies.select_dtypes(include='object')

columns_with_strings = string_columns.columns

print(columns_with_strings)
```

```python
if string_columns.any().any():
    print("There are columns with string values.")
else:
    print("There are no columns with string values.")
```

```python
train_dummies.to_csv('Data/cleaned_train.csv', index=False, mode='w')
```
